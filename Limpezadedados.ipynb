Com o numero grande de ausencia de idades, podemos usar alguma das opçoes, tirando uma media ou mediana das mesmas.

```python
# Visualizacao em boxplot
titanic.boxplot(column='Age')
```

```python
# Visualizacao em histograma
titanic.Age.plot.hist(bins=20);
```

```python
# Se a escolha for substituir a idade faltante por a mediana
titanic.Age.median()
```

```python
# Ou pela mediana tirada por algum grupo
titanic.groupby('Pclass').Age.median()
```

```python
# Ou pela media tirada por alguns grupos
titanic.groupby(['Pclass', 'Sex']).Age.median()
```

Com a mediana das idades por classe e sexo, pode ser criada uma nova coluna para ajustar idades.

```python
# Criando uma nova coluna de idades
titanic['Age_median'] = titanic.groupby(['Pclass', 'Sex']).Age.transform('median')
```

```python
# Visualizando a nova coluna
titanic.head(20)
```

Agora pode ser usada a coluna Age_median, para filtrar por sexo e classe as celulas de idade vazias e substituir pela mediana criada.

```python
# Usando a coluna Age_median para criar a coluna Age_new
titanic['Age_new'] = titanic.Age.fillna(titanic.groupby(['Pclass', 'Sex']).Age.transform('median'))
```

```python
# Visualizando a nova coluna
titanic.head(10)
```

Apos criar as colunas e verificar as idades podemos substituir as colunas vazias pelos dados da coluna Age_new.

```python
# Substituindo as celulas de idades vazias
titanic['Age'] = titanic.Age.fillna(titanic.groupby(['Pclass', 'Sex']).Age.transform('median'))
```

```python
# Visualizando tabela atualizada
titanic.head(3)
```

```python
# Eliminando as colunas Age_median	e Age_new.
titanic = titanic.drop(['Age_new','Age_median'],axis=1)
```

```python
# Visualizando tabela sem as colunas Age_median	e Age_new.
titanic.head(3)
```

```python
# Verificando valores nulos
titanic.isnull().sum()
```
